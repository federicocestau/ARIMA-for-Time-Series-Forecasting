Let’s merge “df_industry” and “df_company_info” first.

We are going to use the column IndustryId in the main DataFrame, to look up the associated columns (data) from the other table. In this case, our main DataFrame should be “df_company_info”, which has duplicate IndustryIds for the companies in the same industry. On the other hand, “df_industry” has a list of unique IndustryIds, which is used to look for the Sectors and Industries. Let’s do the following to merge these two tables.

```python
#it agregates from the right dataframe the columns the other one does not have. 
#All left dataframe rows are going tyo be retrieve. 
# So, all rows within the main DataFrame (the left one) will be kept and we will only look for the data from the right DataFrame according to the given column/columns.
df_company_info_all = pd.merge(left = df_company_info,right=df_industry,how='left',on='IndustryId')
df_company_info_all 
```

Let’s start with counting how many stocks we have within the sectors. It is important because to calculate Z scores, we need two stocks at least. From the financial analysis point of view, we may need more than ten stocks.

```python
df_profit_sector_group = df_profit_sector.groupby('Sector').count()
df_profit_sector_group
```

Sector level Z scores

Z-score is a statistical measurement that describes a value's relationship to the mean of a group of values. Z-score is measured in terms of standard deviations from the mean. If a Z-score is 0, it indicates that the data point's score is identical to the mean score. A Z-score of 1.0 would indicate a value that is one standard deviation from the mean. Z-scores may be positive or negative, with a positive value indicating the score is above the mean and a negative score indicating it is below the mean.

We will use the function “merge”, but before that, we need to use the function “reset_index” to make the index column, SECTOR, as a regular column.

```python
df_profit_sector_mean = df_profit_sector_mean.reset_index()
df_profit_sector_mean
```

```python
df_profit_sector_std = df_profit_sector_std.reset_index()
df_profit_sector_std
```

```python
df_profit_sector_sub.head(5)
```

```python
#it includes the new columns from the left. The ones that are the same are not included. 
df_profit_sector_sub = pd.merge(left=df_profit_sector_sub,
                                right=df_profit_sector_mean,
                                how='left',
                                on='Sector',
                                suffixes = ['','_MEAN']
                                )
```

```python
df_profit_sector_sub
```

```python
df_profit_sector_sub = pd.merge(left=df_profit_sector_sub,
                                right=df_profit_sector_std,
                                how='left',
                                on='Sector',
                                suffixes = ['','_STD']
                                )
```

```python
df_profit_sector_sub.columns
```

```python
df_profit_sector_sub.head()
```

```python
df_profit_sector_sub['Revenue_Z'] = (df_profit_sector_sub['Revenue'] - df_profit_sector_sub['Revenue_MEAN'])/df_profit_sector_sub['Revenue_STD']
```

```python
df_profit_sector_sub['Gross Profit_Z'] = (df_profit_sector_sub['Gross Profit'] - df_profit_sector_sub['Gross Profit_MEAN'])/df_profit_sector_sub['Gross Profit_STD']
```

```python
df_profit_sector_sub['Net Income_Z'] = (df_profit_sector_sub['Net Income'] - df_profit_sector_sub['Net Income_MEAN'])/df_profit_sector_sub['Net Income_STD']
```

```python
df_profit_sector_sub['Z'] = (df_profit_sector_sub.Revenue_Z + df_profit_sector_sub['Gross Profit_Z'] + df_profit_sector_sub['Net Income_Z'])/3
```

```python
df_top_10 = df_profit_sector_sub.sort_values('Z',ascending=False).head(10)
```

```python
df_top_10
```

Rolling
Some time-series computation is based on a rolling window. For example, if we need a simple moving average, we need to take a window (a sub-series) to calculate the average, then roll it down with the fixed size of the window. We can easily achieve so with the function “rolling”.

Here we used the function “rolling” with the column “close”. We set the window size to be 5, so it will take the first 5 days to calculate the average with the function “mean”, then move down one day with another 5 days window. We then created a new column “SMA5”.

```python
#every 5 time period, you calculate the mean. 
spy_df['Mean5DF'] = spy_df.Close.rolling(5).mean()
```

The exponential moving average is the base for many other indicators, Let’s do one example to calculate 20 days exponential moving average.

```python
df = pd.read_csv('https://raw.githubusercontent.com/flyandlure/datasets/master/ecommerce_sales_by_date.csv', 
                usecols=['date', 'transactions'],
                parse_dates=['date'])
df.head(10)
```

Calculate a simple moving average

```python
df = df.assign(sma_3day = df['transactions'].rolling(3).mean())
df.head()
```

EWMA is used to calculate the day-to-day fluctuations in accounting/financial processes which may seem to be very large, it can be used to calculate the fluctuations of an e-commerce website or other website, etc.,

Pandas ema function: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.ewm.html

For ewm() function comass, span, halflife, and alpha are mutually exclusive, so to plot ewm().mean() for different alpha values, span cannot be provided. When the span is provided alpha is calculated as 2/(span+1) where span >= 1.

https://medium.com/codex/simple-moving-average-and-exponentially-weighted-moving-average-with-pandas-57d4a457d363

# Calculate an exponential moving average

```python
df = df.assign(
    ewm_3day = df['transactions'].ewm(span=3, adjust=False).mean(),    
    ewm_7day = df['transactions'].ewm(span=7, adjust=False).mean(),
    ewm_10day = df['transactions'].ewm(span=7, adjust=False).mean(),
)

df.head(10)
```

```python
df = df.assign(
    ewm_3day_Adjust = df['transactions'].ewm(span=3, adjust=True).mean(),    
    ewm_7day_alpha = df['transactions'].ewm(alpha=0.4).mean()
)
df.head()
```

As the value of alpha increases, it becomes more and more similar to the original feature.

```python
# setting first name as index column
df.set_index('date', inplace = True)
```
